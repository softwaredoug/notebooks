## Get dot n dot products

Get an array of dot products from random vectors of the provided dimensions

```python
def dots_of_n(dims, rounds=1000):
    dots = []
    for _ in range(rounds):
        v1 = random_vect(dims)
        v2 = random_vect(dims)
        dotted = np.dot(v1, v2)
        dots.append(dotted)
        
    return np.array(dots)
        
        
dots_of_n(10, rounds=10)
```

## Plots at n-dimensions

As dimensions increase, we see the variance of the distribution shrinks.

```python
from matplotlib import pyplot as plt 

dots = dots_of_n(10)

plt.title('Dot product, 10 dimensions')
plt.hist(dots, bins=np.arange(-1,1,0.01))
plt.show()
```
